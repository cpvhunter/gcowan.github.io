# Root finding

The problem is to find the values of x where f(x)=0. It's a good idea to have an idea of what f(x) looks like before starting, so that you can make a good initial guess at the region of interest that contains the root (or roots). There are a few different methods that we can discuss. Let's first try using this function as an example.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='serif', size=80)
import pylab
%matplotlib inline
pylab.rcParams['figure.figsize'] = 8, 8
```

```python
f = lambda x: 10 - 10*x - 2*x**2 + x**3
x_vals = np.arange(-4, 5, 0.1)
fig, ax = plt.subplots()
ax.plot(x_vals, f(x_vals), 'b-', label='f(x)')
pylab.grid(True)
legend = ax.legend(loc='upper center')
ax.set_ylabel('f(x)',fontsize=20)
ax.set_xlabel('x',fontsize=20)
```

## Bisection method

Here, we determine the range where the root is known to exist. It's clear that a root must exist in some range as f(x) will have a different sign at the minimum and maximum values of this range. The algortihm proceeds by evaluating f(x) at the mid-point of the range and comparing the sign of the value to the signs of f(x) at the minimum and maximum of the range. We then make a new range from the mid-point and either the old min or max values of x. The choice of which to use is done based upon the signs of f(x) at each point. Simply repeat this procedure (halving the size of the range at each iteration) until the desired accuracy is reached. We can define accuracy as meaning the difference between the min and max values of the range.

```python
def bisection(f, x_min, x_max, tolerance):
    count = 0
    while x_max - x_min > tolerance:
        count += 1
        x_mid = x_min + (x_max - x_min)/2.
        f_min = f(x_min)
        f_max = f(x_max)
        f_mid = f(x_mid)
        #print x_min, x_mid, x_max
        # Check if the function changes sign between the minimum and the mid-point.
        # If it stays the same, then move the new minimum to the current mid-point.
        # If not, then the root must be between the minimum and mid-point, so
        # move the current maximum to the current mid-point. Then repeat.
        if f_mid/f_min > 0:
            x_min = x_mid
        else:
            x_max = x_mid
    print count
    return x_mid
```

```python
tolerance = 0.0000001
roots = np.zeros(3)
roots[0] = bisection(f, -5,  0, tolerance)
roots[1] = bisection(f,  0,  3, tolerance)
roots[2] = bisection(f,  1,  5, tolerance)
print roots
```

```python
fig, ax = plt.subplots()
ax.plot(x_vals, f(x_vals), 'b-', label='f(x)')
ax.plot(roots, f(roots), 'ro', label='roots Bi')
pylab.grid(True)
legend = ax.legend(loc='upper center')
ax.set_ylabel('f(x)',fontsize=20)
ax.set_xlabel('x',fontsize=20)
```

This is an example of a method that cannot fail. However, there are a few features that you need to be aware of.

   * You know in advance the number of steps needed to a given accuracy
   * It may, however, not converge on a root. e.g. for f(x) = 1/(x-a)  it will converge on “a” 
   * It may take longer than other methods which is important in major code bases
   * It only works with continuous functions. 
   * It only requires evaluation of f(x) (which is advantageous when the function is something that comes out of a large calculation)
   * Works well for 1D, inefficient for multiple dimensions.

## Newton-Raphson method

This method is more complex as we use information about the derivative of f(x) in order to refine the estimate of the root and converge on it more quickly. It can be explained as follows. First, Taylor expand f(x) around some initial guess for the root ($x_0$).

$f(x_0 + d) = f(x_0) + d\, f^\prime(x_0) + \frac{d^2}{2}\, f^{\prime\prime}(x_0) + \dots$

Your first guess will probably not be correct, so $f(x_0) \neq 0$. But if you are near the root then you can write (ignoring second order terms)

$f(x_0+d) = 0 = f(x_0) + d\, f^\prime(x_0)$

If you then solve for $d$

$d = -f(x_0)/f^\prime(x_0)$

So you make $x_0 + d$ your next guess for where the root is and iterate the process.

```python
def newton_raphson(f, df, x0, tolerance):
    count = 0
    while np.abs(f(x0)) > tolerance:
        count += 1
        d = -f(x0)/float(df(x0))
        x0 = x0 + d
    print count
    return x0
```

```python
# Define the derivative of our example function
df = lambda x: -10 - 4*x + 3*x**2
```

```python
roots_nr = np.zeros(3)
roots_nr[0] = newton_raphson(f, df, 2, tolerance)
roots_nr[1] = newton_raphson(f, df, -2, tolerance)
roots_nr[2] = newton_raphson(f, df, 5, tolerance)
print roots_nr
print roots
```

```python
fig, ax = plt.subplots()
ax.plot(x_vals, f(x_vals), 'b-', label='f(x)')
ax.plot(roots_nr, f(roots_nr), 'ro', label='roots NR')
pylab.grid(True)
legend = ax.legend(loc='upper center')
ax.set_ylabel('f(x)',fontsize=20)
ax.set_xlabel('x',fontsize=20)
```

As you can see from the counters in each method, the Newton-Raphson method is much faster to converge to the solution. However, the method itself is unstable and, depending on the starting value that is given it may not converge.
