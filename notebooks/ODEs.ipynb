# Solving first order ordinary differential equations (ODEs)

A general first order ODE can be expressed as $\frac{dy(x)}{dx} = f(x, y)$. For example: $\frac{dy(x)}{dx} = y(x) + x + 2x^2$. Typically, a physics problem may involve higher-order differential equations, but these can always be reduced to studying a set of coupled first order equations, i.e., $\frac{dy_i(x)}{dx} = f_i(x, y_0, \dots, y_{N-1})$, where the functions $f_i$ on the RHS are known.

Solving the ODE for the form of the $y_i(x)$ functions may be possible analytically. However, where this is not possible, numerical methods can be used. In this example we use some numerical methods to solve first order ODEs, showing some differences in accuracy of the approaches. Knowledge of the ODE iteself is not sufficient to solve it, we must also know the boundary conditions, which define that value of the functions at specific points.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='serif')
import pylab
%matplotlib inline
pylab.rcParams['figure.figsize'] = 8, 8
```

Here we define the differential equation (the function $f$ on the RHS).

```python
f = lambda x, y: 1 - 6*x + 3*x**2
```

```python
def euler(f, x0, y0, steps, xmin, xmax):
    """Implements the Euler method for solving an ODE
    for the function y between xmin and xmax,
    given initial conditions y(x0) = y0
    """
    x = [x0]
    y = [y0]
    delta = float(xmax - xmin)/steps
    for i in range(1, steps+1):
        y.append(y[-1] + delta*f(x[-1] + delta, y[-1]))
        x.append(x[-1] + delta)
    return x, y
```

Run the algorithm.

```python
x_num_euler1, y_num_euler1 = euler(f, 0, 1, 40, 0, 4)
```

This is the analytic soltuion to the ODE.

```python
x_data1 = np.arange(0,4,0.1)
y = lambda x: 1 + x - 3*x**2 + x**3
```

Now make a plot of the analytical function and our numerical solution.

```python
fig, ax = plt.subplots()
ax.plot(x_data1, y(x_data1), 'b-', label='Analytic solution')
ax.plot(x_num_euler1, y_num_euler1, 'ro', label='Numerical solution (Euler)')
legend = ax.legend(loc='upper center')
ax.set_ylabel('y(x)')
ax.set_xlabel('x')
```

Now try a different ODE and repeat the process.

```python
f2 = lambda x, y: np.cos(x)
```

```python
x_num_euler2, y_num_euler2 = euler(f2, 0, 0, 40, 0, 20)
```

```python
x_data2 = np.arange(0,20,0.1)
y2 = lambda x: np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x_data2, y2(x_data2), 'b-', label='Analytic solution')
ax.plot(x_num_euler2, y_num_euler2, 'ro', label='Numerical solution (Euler)')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)')
ax.set_xlabel('x')
```

It's clear that for too small steps, it becomes difficult to follow the exact form of the curve. Now let's implement the midpoint Runge-Kutta method (or second order RK), where we take the gradient to the curve at the midpoint of range in each iteration and then use that gradient when extrapolating from the start of the range.

```python
def RK2(f, x0, y0, steps, xmin, xmax):
    """Implements the midpoint RK method for solving an ODE
    for the function y between xmin and xmax,
    given initial conditions y(x0) = y0
    """
    x = [x0]
    y = [y0]
    y_mid = 0.
    delta = float(xmax - xmin)/steps
    for i in range(1, steps+1):
        k1 = delta*f(x[-1], y[-1])
        k2 = delta*f(x[-1] + delta/2., y[-1] + k1/2.)
        y.append(y[-1] + k2)
        x.append(x[-1] + delta)
    return x, y
```

```python
x_num_RK2, y_num_RK2 = RK2(f2, 0, 0, 40, 0, 20)
```

```python
fig, ax = plt.subplots()
ax.plot(x_data2, y2(x_data2), 'b-', label='Analytic solution')
ax.plot(x_num_euler2, y_num_euler2, 'ro', label='Numerical solution (Euler)')
ax.plot(x_num_RK2, y_num_RK2, 'yo', label='Numerical solution (RK2)')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)')
ax.set_xlabel('x')
```

Now let's use a non-linear ODE that has some y depdenence (the above don't have y on the RHS).

```python
f3 = lambda x, y: 1/y
```

```python
x_num_euler3, y_num_euler3 = euler(f3, 0, 1, 40, 1, 10)
x_num_RK2_2, y_num_RK2_2 = RK2(f3, 0, 1, 40, 1, 10)
```

The analytical solution here is just $y(x) = \sqrt{2x+1}$, where $N$ is given by the initial conditions.

```python
x_data3 = np.arange(1,10,0.1)
y3 = lambda x: np.sqrt(2*x + 1)
```

```python
fig, ax = plt.subplots()
ax.plot(x_data3, y3(x_data3), 'b-', label='Analytic solution')
ax.plot(x_num_euler3, y_num_euler3, 'ro', label='Numerical solution (Euler)')
ax.plot(x_num_RK2_2, y_num_RK2_2, 'yo', label='Numerical solution (RK2)')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)')
ax.set_xlabel('x')
```

Automatically we can see that the MRK method it better than the Euler method. Now we can implement somehting even more general, the 4th order Runge-Kutta method. 

```python
def RK4(f, x0, y0, steps, xmin, xmax):
    """Implements the 4th order RK method for solving an ODE
    for the function y between xmin and xmax,
    given initial conditions y(x0) = y0
    """
    x = [x0]
    y = [y0]
    y_mid = 0.
    delta = float(xmax - xmin)/steps
    for i in range(1, steps+1):
        k1 = delta*f(x[-1]           , y[-1])
        k2 = delta*f(x[-1] + delta/2., y[-1] + k1/2.)
        k3 = delta*f(x[-1] + delta/2., y[-1] + k2/2.)
        k4 = delta*f(x[-1] + delta   , y[-1] + k3)
        y.append(y[-1] + k1/6. + k2/3. + k3/3. + k4/6.)
        x.append(x[-1] + delta)
    return x, y
```

Let's try this on the sinusoidal example again.

```python
x_num_RK4, y_num_RK4 = RK4(f2, 0, 0, 40, 0, 20)
```

```python
fig, ax = plt.subplots()
ax.plot(x_data2, y2(x_data2), 'b-', label='Analytic solution')
ax.plot(x_num_euler2, y_num_euler2, 'ro', label='Numerical solution (Euler)')
ax.plot(x_num_RK2, y_num_RK2, 'yo', label='Numerical solution (RK2)')
ax.plot(x_num_RK4, y_num_RK4, 'go', label='Numerical solution (RK4)')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)')
ax.set_xlabel('x')
```

Let's plot the differences of each method with respect to the analytical solution.

```python
fig, ax = plt.subplots()
ax.plot(x_num_euler2, (y_num_euler2 - y2(x_num_euler2)), 'ro', label='Diff. Numerical solution (Euler)')
ax.plot(x_num_RK2, (y_num_RK2 - y2(x_num_RK2)), 'yo', label='Diff. Numerical solution (RK2)')
ax.plot(x_num_RK4, (y_num_RK4 - y2(x_num_RK4)), 'go', label='Diff. Numerical solution (RK4)')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)')
ax.set_xlabel('x')
```

It's clear that the RK4 method is the most accurate.
