# Solving a system of first order ODEs with the fourth order Runge-Kutta method

The previous example that showed how to solve a second order ODE as a set of coupled first order ODEs is a special case of the more general problem of solving a system of coupled first order ODEs. Below, we re-express the RK4 algortihm in matrix notation such that we can generally solve any system of equations.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='serif', size=80)
import pylab
%matplotlib inline
pylab.rcParams['figure.figsize'] = 8, 8
```

We start by expression the problem as $\frac{d\vec{y}}{dx} = \vec{f}(\vec{y(x)}, x)$, where $\vec{y(x)}$ are the functions that were trying to find the solution of and $\vec{f}$ is the vector of derivatives (i.e., the RHS of each separate ODE). Below I specify the derivative vector for the case that we saw previously of the damped and forced harmonic osciallator.

```python
def derivative_vector( solution_vector, x ):
    k = 10
    m = 1
    c = 2
    F0 = 2
    omega = 1.5
    F = lambda x: F0*np.sin(omega*x)
    f0prime = lambda x, y0, y1: y1
    f1prime = lambda x, y0, y1: -k/m*y0 - c/m*y1 + F(x)
    f = np.zeros(2)
    y0 = solution_vector[0]
    y1 = solution_vector[1]
    f[0] = f0prime(x, y0, y1)
    f[1] = f1prime(x, y0, y1)
    return f
```

Here is the RK4 algorithm, expressed in vector notation. It is quite simple when viewed like this.

```python
def RK4(derivative_vector, x_0, y_0, steps, xmin, xmax):
    """Implements the 4th order RK method for solving a
    system of linear ODEs, specified by the derivative vector,
    for the \vec{y} between xmin and xmax,
    given initial conditions \vec{y(x0)} = \vec{y_0}.
    """
    x = np.zeros(steps)
    y = np.zeros((steps, len(y_0)))
    x[0] = x_0
    y[0] = y_0
    delta = float(xmax - xmin)/steps
    for i in range(1, steps):
        k1 = delta*derivative_vector( y[i-1]         , x[i-1]           )
        k2 = delta*derivative_vector( y[i-1] + 0.5*k1, x[i-1] + delta/2.)
        k3 = delta*derivative_vector( y[i-1] + 0.5*k2, x[i-1] + delta/2.)
        k4 = delta*derivative_vector( y[i-1] +     k3, x[i-1] + delta   )
        x[i] = x[i-1] + delta
        y[i] = y[i-1] + 1/6.*(k1 + 2.*k2 + 2.*k3 + k4)
    return x, y
```

Now let's solve it and we should get the same answer we had previously.

```python
x_num_RK4, y_num_RK4 = RK4(derivative_vector, 0, np.array([1, 0]), 400, 0, 5)
```

```python
fig, ax = plt.subplots()
ax.plot(x_num_RK4, y_num_RK4[:,0], 'ro', label='y0 solution (damped + forced)')
ax.plot(x_num_RK4, y_num_RK4[:,1], 'bo', label='y1 solution (damped + forced)')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)', fontsize=20)
ax.set_xlabel('x', fontsize=20)
```

Let's try a simple 1D example to show the vector code can deal with this.

```python
def derivative_vector2( solution_vector, x ):
    f0prime = lambda x, y0: 1/y0
    f = np.zeros(1)
    y0 = solution_vector[0]
    f[0] = f0prime(x, y0)
    return f
```

```python
x_num_RK4, y_num_RK4 = RK4(derivative_vector2, 0, np.array([1]), 400, 0, 5)
```

```python
fig, ax = plt.subplots()
ax.plot(x_num_RK4, y_num_RK4[:,0], 'ro', label='y0 solution')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)', fontsize=20)
ax.set_xlabel('x', fontsize=20)
```

Here is another example with initial conditions $y_0(0) = -1, y_1(0) = 0$

$y_0^\prime =  y_0 - y_1 + 2$,

$y_1^\prime = -y_0 + y_1 + 4x$.

The exact solution here is

$y_0(x) = -\frac{1}{2}e^{2x} + x^2 + 2x - \frac{1}{2}$,

$y_1(x) =  \frac{1}{2}e^{2x} + x^2      - \frac{1}{2}$

```python
def derivative_vector3( solution_vector, x ):
    f0prime = lambda x, y0, y1: y0 - y1 + 2
    f1prime = lambda x, y0, y1: -y0 + y1 + 4*x
    f = np.zeros(2)
    y0 = solution_vector[0]
    y1 = solution_vector[1]
    f[0] = f0prime(x, y0, y1)
    f[1] = f1prime(x, y0, y1)
    return f
```

```python
x_num_RK4, y_num_RK4 = RK4(derivative_vector3, 0, np.array([-1, 0]), 50, 0, 2)
```

```python
fig, ax = plt.subplots()
ax.plot(x_num_RK4, y_num_RK4[:,0], 'ro', label='y0 solution')
ax.plot(x_num_RK4, y_num_RK4[:,1], 'bo', label='y1 solution')
ax.plot(x_num_RK4, -0.5*np.exp(2*x_num_RK4) + x_num_RK4**2 + 2*x_num_RK4 - 0.5, 'r-', label='y0 analytic')
ax.plot(x_num_RK4, +0.5*np.exp(2*x_num_RK4) + x_num_RK4**2               - 0.5, 'b-', label='y1 analytic')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)', fontsize=20)
ax.set_xlabel('x', fontsize=20)
```
