# Statistics: Example of using the bootstrap method to determine the accuracy of an estimator of a distribution

Bootstrapping is a method that can be used to estimate the accuracy of an estimator of a distribution. It refers to any method that relies on random sampling with replacement. For example, the estimator that you are interested in could be the mean of a distribution. Bootstrapping allows us to calculate the uncertainty on that mean. There is a lot of online information about this, what follows is just a simple example where we generate N events from a Gaussian distribution with mean $\mu$ and width $\sigma$ and use boostrapping to get the uncertainty on the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='serif', size=80)
import pylab
%matplotlib inline
pylab.rcParams['figure.figsize'] = 8, 8
```

```python
mu, sigma = 2.5, 0.7 # mean and standard deviation
s = np.random.normal(mu, sigma, 10000)
count, bins, ignored = plt.hist(s, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
```

Now compute the mean and standard deviation for this sample, these are estimators of the real underlying distribution.

```python
mean = np.mean(s)
std_dev = np.std(s)
mean, std_dev
```

Define the bootstrap method to compute the uncertainty on these quantities. For each iteration, we want to create a new array of 1000 entries that have been randomly sampled from the array s, *with replacement*. This means that some of the events in the array s will appear more than once in the new array.

```python
def bootstrap( s, f, iterations ):
    """f is the function that computes the estimator of interest
    """
    N = len(s)
    estimators = np.zeros(iterations)
    for i in range(iterations):
        new_s = np.zeros(N)
        for n in range(N):
            index = np.random.uniform(0, N)
            new_s[n] = s[index]
        estimators[i] = f(new_s)
    return estimators
```

```python
std_devs = bootstrap( s, np.std , 1000 )
means    = bootstrap( s, np.mean, 1000 )
```

The standard deviation of these distributions can be taken as a estimate of the uncertainty on the mean and standard deviation of the original distribtion of interest.

```python
count2, bins2, ignored2 = plt.hist(std_devs, 30, normed=True)
```

```python
count3, bins3, ignored3 = plt.hist(means, 30, normed=True)
```

So the final mean and standard deviation of the Gaussian of 10000 events are

```python
print "mean = %0.3f +- %0.3f" % (mean, np.std(means))
print "std dev = %0.3f +- %0.3f" % (std_dev, np.std(std_devs))
```

There are, of course, simpler ways of computing these quantities in this simplified example, but it demonstrates the method.
