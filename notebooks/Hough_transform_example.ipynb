# Hough transforms (Ex 1)

Hough transforms are a way of representing an image such that the parameters describing arbitrary shapes (typically lines or circles) in the image can be easily determined. The image need not be 2D. This technique is sometimes used in particle physics experiments to determine the parameters of particle tracks, where the image is the map of "hits" that the particle leaves as is traverses the detector medium (i.e., such as ionisation trail in a silicon sensor).

First import the required packages that we need.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

Define a function that will be used to generate some random tracks in our 2D parameter space.

```python
def generateTrack( nHits ):
    theta = random.uniform(-np.pi, np.pi)
    rho   = random.uniform(-100*np.sqrt(2.), 100*np.sqrt(2.))
    #rho   = random.uniform(0., 10.)
    tmpx = random.sample(xrange(100), nHits)
    x = []
    y = []
    for i in tmpx:
        tmpy = (rho - i*np.cos(theta))/np.sin(theta)
        if tmpy < 100 and tmpy > 0:
            x.append(i)
            y.append(tmpy)
    return (x, y, rho, theta)
```

Now create some tracks, making sure that we have at least 3 points on each.

```python
tracks = []
while len(tracks) < 3:
    x, y, rho_gen, theta_gen = generateTrack( 7 )
    #x, y, rho_gen, theta_gen = generateRing( 100 )
    if len(y) > 5:
        print x, y, rho_gen, theta_gen
        tracks.append((x,y))
```

Now define the parameters of the $(\rho, \theta)$ space (the Hough space).

```python
dim = 100
theta_min = -np.pi
theta_max = +np.pi
rho_min = -100.*np.sqrt(2.)
rho_max = 100.*np.sqrt(2.)
```

The accumulator is a 2D array that is used to keep track of where the lines intersect in the $(\rho, \theta)$ space.

```python
bins = 200
theta_list = np.linspace(theta_min, theta_max, num=bins, endpoint=False)
accumulator = np.zeros((bins, bins))
maxima = []
```

Now loop over all of the points in each track and follow their position through the Hough space. Keep track of the maxima, so that we can apply a sensible threshold at the end.

```python
for track in tracks:
    print "track", track
    for i in range(len(track[0])):
        x = track[0][i]
        y = track[1][i]
        for theta in theta_list:
                rho = x*np.cos(theta) + y*np.sin(theta)
                #rho = np.sqrt((x-theta)**2 + y*y)
                rho_bin   = int (bins*(rho-rho_min)/(rho_max-rho_min) );
                theta_bin = int (bins*(theta-theta_min)/(theta_max-theta_min) );
                #print rho, rho_bin, rho_max, theta, theta_bin
                accumulator[theta_bin, rho_bin] += 1
    maxima.append(accumulator.max())
```

Now only look for the intersections of the lines in Hough space by applying a threshold cut. The points in the accumulator that survive this cut have the $(\rho, \theta)$ coordinates that correspond to the parameters of the tracks in the real space.

```python
indices = []
for i, val in enumerate(accumulator >= 0.75*min(maxima) ):
    #print i, val
    count = 0
    for v in val:
        if v:
            indices.append((i, count))
        count += 1
print indices
```

```python
plt.imshow(accumulator, origin='lower', extent=[rho_min, rho_max, theta_min, theta_max], aspect='auto')
plt.show()
```

Now use the solutions to create some lines that we can draw on top of the tracks that we generated at the start.

```python
x_list = np.arange(0, 100)
lines = []
for theta_bin, rho_bin in indices:
    rho   = rho_min   + rho_bin  *(rho_max-rho_min)/bins
    theta = theta_min + theta_bin*(theta_max-theta_min)/bins
    print "rho", rho, "rho_min", rho_min, "rho_max", rho_max, "rho_bin", rho_bin, "comp", rho_bin  *(rho_max-rho_min)/bins, theta, theta_bin
    y = [ (rho - i*np.cos(theta))/np.sin(theta) for i in x_list]
    lines.append(y)
```

Plot the tracks that we generated and the lines corresponding to the solutions.

```python
for track in tracks:
    plt.scatter(track[0], track[1])
for line in lines:
    plt.plot(x_list, line)
#axarr[1].plot(x_list, y_gen, c='r')
```
