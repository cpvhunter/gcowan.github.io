# Solving second order ODEs with the fourth order Runge-Kutta method

A common approach for solving a second order ODE with the Runge-Kutta (or any numerical method) is to split the ODE into two coupled first order ODEs and then to run an RK4 algorithm that simultaneously solves both coupled equations.

## Simple harmonic oscillator

For example, consider the textbook case of simple harmonic motion, where a spring moves in the $y$ direction as a function of time, $x$. The spring has a spring constant, $k$, and a mass, $m$, is attached to it. The second order ODE takes the form

$\frac{d^2y}{dx^2} = -\frac{k}{m} y$.

We want to solve for $y(x)$. The solution is to split this into two parts

$\frac{dz}{dx} = -\frac{k}{m} y$, $\frac{dy}{dx} = z$.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='serif', size=80)
import pylab
%matplotlib inline
pylab.rcParams['figure.figsize'] = 8, 8
```

```python
k = 10
m = 1
zprime = lambda x, y, z: -k/m*y
yprime = lambda x, y, z: z
```

```python
def RK4_SecondOrder(yprime, zprime, x0, z0, y0, steps, xmin, xmax):
    """Implements the 4th order RK method for solving a
    second order ODE for the function y between xmin and xmax,
    given initial conditions y(x0) = y0, y'(x0) = z(x0) = z0
    """
    x = [x0]
    y = [y0]
    z = [z0]
    delta = float(xmax - xmin)/steps
    for i in range(1, steps+1):
        k1 = delta*yprime(x[-1]           , y[-1]        , z[-1])
        j1 = delta*zprime(x[-1]           , y[-1]        , z[-1])
        k2 = delta*yprime(x[-1] + delta/2., y[-1] + k1/2., z[-1] + j1/2.)
        j2 = delta*zprime(x[-1] + delta/2., y[-1] + k1/2., z[-1] + j1/2.)
        k3 = delta*yprime(x[-1] + delta/2., y[-1] + k2/2., z[-1] + j2/2.)
        j3 = delta*zprime(x[-1] + delta/2., y[-1] + k2/2., z[-1] + j2/2.)
        k4 = delta*yprime(x[-1] + delta   , y[-1] + k3   , z[-1] + j3)
        j4 = delta*zprime(x[-1] + delta   , y[-1] + k3   , z[-1] + j3)
        x.append(x[-1] + delta)
        y.append(y[-1] + 1/6.*(k1 + 2.*k2 + 2.*k3 + k4))
        z.append(z[-1] + 1/6.*(j1 + 2.*j2 + 2.*j3 + j4))
    return x, y
```

Let's solve the equation, with the initial conditions y(0) = 1, y'(0) = z(0) = 0.

```python
x_num_RK4, y_num_RK4 = RK4_SecondOrder(yprime, zprime, 0, 0, 1, 400, 0, 5)
```

```python
fig, ax = plt.subplots()
ax.plot(x_num_RK4, y_num_RK4, 'go', label='Numerical solution (RK4) simple')
legend = ax.legend(loc='lower center')
ax.set_ylabel('y(x)', fontsize=20)
ax.set_xlabel('x', fontsize=20)
```

The above shows the expected sinusoidal behaviour. 

## Damped oscillator

Now let's introduce a damping term with coefficient $c$, so the equations become

$\frac{d^2y}{dx^2} = -\frac{k}{m} y - \frac{c}{m}\frac{dy}{dx}$.

Again, split into two coupled equations

$\frac{dz}{dx} = -\frac{k}{m} y - \frac{c}{m}z$, $\frac{dy}{dx} = z$.

```python
c = 2
zprime2 = lambda x, y, z: -k/m*y - c/m*z
yprime2 = lambda x, y, z: z
```

```python
x_num_RK4_2, y_num_RK4_2 = RK4_SecondOrder(yprime2, zprime2, 0, 0, 1, 400, 0, 5)
fig, ax = plt.subplots()
ax.plot(x_num_RK4_2, y_num_RK4_2, 'go', label='Numerical solution (RK4) damped')
legend = ax.legend(loc='upper center')
ax.set_ylabel('y(x)', fontsize=20)
ax.set_xlabel('x', fontsize=20)
```

## Forced harmonic oscillator 

Now introduce a driving force, $F(x)$, to the oscillator, so the equation becomes

$\frac{d^2y}{dx^2} = -\frac{k}{m} y - \frac{c}{m}\frac{dy}{dx} + F(x)$.

Again, split into two coupled equations

$\frac{dz}{dx} = -\frac{k}{m} y - \frac{c}{m}z + \frac{1}{m}F(x)$, $\frac{dy}{dx} = z$.

Consider the case where $F(x) = F_0\sin(\omega x)$ where $F_0$ is the driving amplitude and $\omega$ is the driving frequency.

```python
F0 = 2
omega = 1.5
F = lambda x: F0*np.sin(omega*x)
zprime3 = lambda x, y, z: -k/m*y - c/m*z + F(x)
yprime3 = lambda x, y, z: z
```

```python
x_num_RK4_3, y_num_RK4_3 = RK4_SecondOrder(yprime3, zprime3, 0, 0, 1, 400, 0, 5)
fig, ax = plt.subplots()
ax.plot(x_num_RK4_2, y_num_RK4_2, 'go', label='Numerical solution (RK4) damped')
ax.plot(x_num_RK4_3, y_num_RK4_3, 'ro', label='Numerical solution (RK4) damped + forced')
legend = ax.legend(loc='upper center')
ax.set_ylabel('y(x)', fontsize=20)
ax.set_xlabel('x', fontsize=20)
```
